```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# normal distribution parameter
sigma = 0.1
mu = [0.3, -0.8, 1.4, 0.4, 1.0, -1.5, -0.2, -1.1, 0.9, -0.9]

def bandit(movemnt):
    return np.random.normal(mu[movemnt], sigma, 1)
```

```python
# parameter for system
# step 1: initialize
iter_num = 1000
Q = [0] * 10
N = [0] * 10
R = [0] * iter_num      # record each reward
epson = 0.1

for i in range(iter_num):
    if random.random() < (1 - epson):
        movement = Q.index(max(Q))
    else:
        movement = random.randint(0, 9)
    reward = bandit(movement)
    if i != 0:
        R[i] = (R[i - 1] * (i - 1) + reward) / i
    else:
        R[i] = reward
    N[movement] += 1
    Q[movement] += (reward - Q[movement]) * 0.1

plt.ylim((0, 1.5))
plt.xlim((0, iter_num))
plt.plot(range(iter_num), R)
plt.show()
print(R[iter_num - 1])
```
